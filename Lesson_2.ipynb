1. Создайте мешок слов с помощью sklearn.feature_extraction.text.CountVectorizer.fit_transform(). Применим его к 'tweet_stemmed' и tweet_lemmatized' отдельно.

Игнорируем слова, частота которых в документе строго превышает порог 0.9 с помощью max_df.

Ограничим количество слов, попадающий в мешок, с помощью max_features = 1000.
Исключим стоп-слова с помощью stop_words='english'. 

Отобразим Bag-of-Words модель как DataFrame. columns необходимо извлечь с помощью CountVectorizer.get_feature_names().


```python
twd = TreebankWordDetokenizer()
doc_stemmed = data['tweet_stemmed'].apply(lambda x: twd.detokenize(x))
doc_stemmed
```

```python
count_vectorizer = CountVectorizer(ngram_range=(1, 1), analyzer='word', binary=True, max_df=0.9, max_features = 1000)

# Создаем the Bag-of-Words модель
count_bag_of_words_stemmed = count_vectorizer.fit_transform(doc_stemmed)

# Отобразим Bag-of-Words модель как DataFrame
feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(count_bag_of_words_stemmed.toarray(), columns = feature_names)
```

```python
doc_lemmatized = data['tweet_lemmatized'].apply(lambda x: twd.detokenize(x))
doc_lemmatized
```

```python
count_vectorizer = CountVectorizer(ngram_range=(1, 1), analyzer='word', binary=True, max_df=0.9, max_features = 1000)

# Создаем the Bag-of-Words модель
count_bag_of_words_lemmatized = count_vectorizer.fit_transform(doc_lemmatized)

# Отобразим Bag-of-Words модель как DataFrame
feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(count_bag_of_words_lemmatized.toarray(), columns = feature_names)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.9, max_features = 1000, stop_words='english')
tfidf_vector_stemmed = tfidf_vectorizer.fit_transform(doc_stemmed)

# Show the Model as a pandas DataFrame
feature_names = tfidf_vectorizer.get_feature_names()
pd.DataFrame(tfidf_vector_stemmed.toarray(), columns = feature_names)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.9, max_features = 1000, stop_words='english')
tfidf_vector_lemmatized = tfidf_vectorizer.fit_transform(doc_lemmatized)

# Show the Model as a pandas DataFrame
feature_names = tfidf_vectorizer.get_feature_names()
pd.DataFrame(tfidf_vector_lemmatized.toarray(), columns = feature_names)
```

```python
# Загружаем данные
data = open('corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)


xtrain_count =  count_bag_of_words_stemmed.transform(train_x)
xvalid_count =  count_bag_of_words_stemmed.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions = classifier.predict(xvalid_count)
predictions
```
